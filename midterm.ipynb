# Question 1: Forward Pass Computation (20 pts)

```python
print('=' * 60)
print('QUESTION 1: Forward Pass Computation')
print('=' * 60)

x  = np.array([1.0, -0.5])
W1 = np.array([[0.3, -0.4], [0.5, 0.2]])
b1 = np.array([0.1, -0.1])
W2 = np.array([[0.6, -0.3]])
b2 = np.array([0.05])

print('Given:')
print(f'  x  = {x}')
print(f'  W1 =\n{W1}')
print(f'  b1 = {b1}')
print(f'  W2 = {W2}')
print(f'  b2 = {b2}')
```

```python
# (a) Pre-activation: z(1) = W(1)x + b(1)
z1 = W1 @ x + b1

print('(a) Pre-activation values z(1) = W(1)x + b(1):')
print(f'  z1[0] = {W1[0,0]}*({x[0]}) + ({W1[0,1]})*({x[1]}) + {b1[0]}')
print(f'        = {W1[0,0]*x[0]:.4f} + {W1[0,1]*x[1]:.4f} + {b1[0]}')
print(f'        = {z1[0]:.4f}')
print(f'  z1[1] = {W1[1,0]}*({x[0]}) + {W1[1,1]}*({x[1]}) + ({b1[1]})')
print(f'        = {W1[1,0]*x[0]:.4f} + {W1[1,1]*x[1]:.4f} + ({b1[1]})')
print(f'        = {z1[1]:.4f}')
print(f'\n  z(1) = {z1}')
```

```python
# (b) ReLU activation: h = ReLU(z(1))
def relu(z):
    return np.maximum(0, z)

h = relu(z1)

print('(b) ReLU activation h = ReLU(z(1)):')
print(f'  ReLU({z1[0]:.4f}) = max(0, {z1[0]:.4f}) = {h[0]:.4f}')
print(f'  ReLU({z1[1]:.4f}) = max(0, {z1[1]:.4f}) = {h[1]:.4f}')
print(f'\n  h = {h}')
```

```python
# (c) Output: y_hat = sigmoid(W(2)h + b(2))
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z2    = W2 @ h + b2
y_hat = sigmoid(z2)

print('(c) Output computation y_hat = sigma(W(2)h + b(2)):')
print(f'  z(2) = {W2[0,0]}*{h[0]:.4f} + ({W2[0,1]})*{h[1]:.4f} + {b2[0]}')
print(f'       = {W2[0,0]*h[0]:.4f} + {W2[0,1]*h[1]:.4f} + {b2[0]}')
print(f'       = {z2[0]:.4f}')
print(f'\n  y_hat = sigma({z2[0]:.4f})')
print(f'        = 1 / (1 + e^(-{z2[0]:.4f}))')
print(f'        = 1 / (1 + {np.exp(-z2[0]):.6f})')
print(f'        = {y_hat[0]:.6f}')
```

```python
# (d) Binary Cross-Entropy Loss
y_true = 1
y_pred = y_hat[0]
bce    = -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

print('(d) Binary Cross-Entropy (BCE) Loss:')
print(f'  L_BCE = -(y*ln(y_hat) + (1-y)*ln(1-y_hat))')
print(f'        = -(1*ln({y_pred:.6f}) + 0*ln({1-y_pred:.6f}))')
print(f'        = -ln({y_pred:.6f})')
print(f'        = {bce:.6f}')

print('\n' + '=' * 60)
print('QUESTION 1 SUMMARY')
print('=' * 60)
print(f'  z(1)  = {z1}')
print(f'  h     = {h}')
print(f'  y_hat = {y_pred:.6f}')
print(f'  L_BCE = {bce:.6f}')
```

```python
print('=' * 60)
print('QUESTION 2(b): SGD vs Adam')
print('=' * 60)

print('''
--- Why does plain SGD oscillate in a narrow valley? ---

SGD update rule: theta <- theta - lr * grad(L)

In a narrow valley (steep along x, flat along y):
  - The gradient is very large along x, very small along y.
  - With a fixed learning rate:
      * Along x: the step overshoots the minimum -> zigzag oscillations.
      * Along y: the step is far too small -> very slow convergence.
  - Result: the optimizer bounces back and forth across the valley
    wall instead of descending smoothly toward the minimum.

--- How does Adam address this? ---

Adam maintains two running estimates per parameter:

  m_t = beta1 * m_{t-1} + (1 - beta1) * g_t       (1st moment / momentum)
  v_t = beta2 * v_{t-1} + (1 - beta2) * g_t^2     (2nd moment / variance)

Effective update:
  theta <- theta - lr * m_hat_t / (sqrt(v_hat_t) + eps)

Key insight — each parameter gets its own adaptive step size:
  * Along x (large, consistent gradients):
    sqrt(v_t) grows large -> effective step SHRINKS -> oscillations damped.
  * Along y (small gradients):
    sqrt(v_t) stays small -> effective step GROWS -> progress accelerated.

  Combined effect: Adam navigates narrow valleys efficiently
  by independently scaling each dimension.
''')
```

# Question 3: Custom Training Loops with tf.GradientTape

```python
print('=' * 60)
print('QUESTION 3: tf.GradientTape Bug Analysis')
print('=' * 60)

print('''
BUGGY CODE:
-----------
@tf.function
def train_step(model, optimizer, loss_fn, x_batch, y_batch):
    logits = model(x_batch)                                     # Line 1  BUG 1
    loss = loss_fn(y_batch, logits)                             # Line 2  BUG 2
    with tf.GradientTape() as tape:                             # Line 3
        grads = tape.gradient(loss, model.trainable_variables)  # Line 4  BUG 3
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    return loss

BUG 1 & BUG 3 -- Forward pass is outside the tape context
  PROBLEM:
    GradientTape only records operations performed INSIDE its
    context manager. Here, both `logits` (Line 1) and `loss`
    (Line 2) are computed before the tape is opened.
    The tape has no computation graph to differentiate, so
    tape.gradient() returns None -> applying None gradients
    either crashes or silently does nothing.

  FIX:
    Move the entire forward pass and loss computation
    INSIDE the `with tf.GradientTape() as tape:` block.

BUG 2 -- Missing training=True flag
  PROBLEM:
    model(x_batch) is called without training=True.
    Layers such as BatchNormalization and Dropout behave
    differently at train time vs. inference time.
    The default is training=False, so dropout is disabled
    and batch norm uses stored running statistics rather
    than the current batch statistics during training.

  FIX:
    Use model(x_batch, training=True).
''')
```

```python
print('CORRECTED CODE:')
print('-' * 50)

@tf.function
def train_step_corrected(model, optimizer, loss_fn, x_batch, y_batch):
    with tf.GradientTape() as tape:             # tape opened FIRST
        logits = model(x_batch, training=True)  # FIX 1+2: inside tape + training=True
        loss   = loss_fn(y_batch, logits)       # FIX 3: loss also inside tape
    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    return loss

print('''
@tf.function
def train_step_corrected(model, optimizer, loss_fn, x_batch, y_batch):
    with tf.GradientTape() as tape:             # tape opened FIRST
        logits = model(x_batch, training=True)  # FIX 1+2: inside tape + training=True
        loss   = loss_fn(y_batch, logits)       # FIX 3: loss also inside tape
    grads = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    return loss
''')

# Verify with a small test model
print('Verification: running corrected function on a small test model...')
test_model = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation='relu', input_shape=(3,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
opt      = tf.keras.optimizers.Adam()
loss_fn  = tf.keras.losses.BinaryCrossentropy()
x_t = tf.random.normal((8, 3))
y_t = tf.cast(tf.random.uniform((8, 1), maxval=2, dtype=tf.int32), tf.float32)

loss_val = train_step_corrected(test_model, opt, loss_fn, x_t, y_t)
print(f'  Loss computed successfully: {loss_val.numpy():.4f}  (gradients are valid, not None) checked')
```

# Question 4: Data Pipelines and Training Strategies

```python
print('=' * 60)
print('QUESTION 4(a): tf.data Pipeline Order')
print('=' * 60)

print('''
Pipeline A:  .batch(64)     -> .shuffle(10000) -> .prefetch(AUTOTUNE)
Pipeline B:  .shuffle(10000) -> .batch(64)     -> .prefetch(AUTOTUNE)

CORRECT : Pipeline B
WRONG   : Pipeline A

Why is Pipeline A wrong?
  Batches are formed first, then the batches themselves are shuffled.
  The samples WITHIN each batch remain in their original sequential
  order. If the dataset is sorted by class, each batch is dominated
  by a single class, producing biased gradient updates.

Why is Pipeline B correct?
  Individual samples are shuffled before batching, so every batch
  is a random mix of samples from across the full dataset, giving
  unbiased gradient estimates at each step.

Correct ordering:
  1. shuffle()   -- randomise individual samples (large buffer_size)
  2. batch()     -- form batches from shuffled samples
  3. prefetch()  -- overlap CPU preprocessing with GPU computation
''')

x_demo = tf.range(20)
y_demo = tf.range(20)
print('Demonstration with numbers 0-19:')

ds_A = tf.data.Dataset.from_tensor_slices((x_demo, y_demo))
ds_A = ds_A.batch(5).shuffle(20, seed=42)
print('Pipeline A (batch -> shuffle), first batch:', list(ds_A.take(1))[0][0].numpy(), '<- consecutive')

ds_B = tf.data.Dataset.from_tensor_slices((x_demo, y_demo))
ds_B = ds_B.shuffle(20, seed=42).batch(5)
print('Pipeline B (shuffle -> batch), first batch:', list(ds_B.take(1))[0][0].numpy(), '<- truly random')
```

```python
print('=' * 60)
print('QUESTION 4(b): Warmup + Cosine Decay LR Schedule')
print('=' * 60)

warmup_steps  = 1000
lr_max        = 1e-3
lr_min        = 1e-6
total_steps   = 50000

def get_lr(step):
    if step < warmup_steps:
        return lr_max * step / warmup_steps          # linear warmup
    progress = (step - warmup_steps) / (total_steps - warmup_steps)
    return lr_min + (lr_max - lr_min) * 0.5 * (1 + np.cos(np.pi * progress))

lr_500 = get_lr(500)
print(f'\nLearning rate at step 500:')
print(f'  Step 500 < warmup_steps (1000)  ->  linear warmup phase')
print(f'  LR = lr_max * (step / warmup_steps)')
print(f'     = {lr_max} * (500 / 1000)')
print(f'     = {lr_max} * 0.5')
print(f'     = {lr_500:.4e}')

print('''
Why is a warmup phase important?
  At the beginning of training, model weights are randomly
  initialised and gradients are highly noisy with large variance.
  Starting with a large LR immediately would:
    - Cause large, misguided weight updates in the first steps.
    - Destabilise Adam's moment estimates (m_t, v_t begin at zero,
      so early bias-corrected estimates are unreliable).
    - In Transformers, corrupt attention weights before any
      useful signal has been learned.

  Warmup gradually increases LR, letting the optimiser accumulate
  reliable gradient statistics before taking large steps.
''')
```

```python
print('=' * 60)
print('QUESTION 4(c): Gradient Accumulation')
print('=' * 60)

gpu_batch    = 32
target_batch = 256
accum_steps  = target_batch // gpu_batch

print(f'''
GPU batch size      : {gpu_batch}
Target batch size   : {target_batch}
Accumulation steps  : {target_batch} / {gpu_batch} = {accum_steps}

Why {accum_steps} accumulation steps?
  Only {gpu_batch} samples fit in GPU memory at once.
  We run {accum_steps} forward-backward passes, accumulate the
  gradients, and apply them once at the end -- simulating
  a single update from a {target_batch}-sample batch.

Why divide the loss by the number of accumulation steps?
  Without dividing:
    total_grad = g_1 + g_2 + ... + g_8
    This is the SUM of 8 mini-batch gradients.
    It is {accum_steps}x larger than the true mean gradient over
    {target_batch} samples, making the effective LR {accum_steps}x too large.

  With dividing:
    loss_step = loss_fn(y, y_hat) / {accum_steps}   # scale each step
    total_grad = (g_1 + ... + g_8) / 8
    This equals the correct mean gradient -- training is stable.
''')

print('Code example:')
print(f'''
accum_steps = 8
accumulated_grads = [tf.zeros_like(v) for v in model.trainable_variables]

for step in range(accum_steps):
    x_mini, y_mini = get_mini_batch()              # 32 samples
    with tf.GradientTape() as tape:
        logits = model(x_mini, training=True)
        loss   = loss_fn(y_mini, logits) / accum_steps  # <-- normalise!
    grads = tape.gradient(loss, model.trainable_variables)
    accumulated_grads = [ag + g for ag, g in zip(accumulated_grads, grads)]

# Apply once after all accumulation steps
optimizer.apply_gradients(zip(accumulated_grads, model.trainable_variables))
''')
```

# Question 6: Transfer Learning Strategies

```python
print('=' * 60)
print('QUESTION 6(a): Feature Extraction vs Fine-Tuning')
print('=' * 60)

print('''
Feature Extraction (Frozen Base):
  - All base model weights are frozen: base_model.trainable = False
  - Only the newly added classifier head is trained.
  - Pre-trained ImageNet representations (edges, textures, shapes)
    act as a fixed feature extractor for the new task.

Fine-Tuning:
  - The upper layers of the base model are unfrozen.
  - The model is retrained end-to-end with a very small LR.
  - The network adapts its features to the target domain.

Recommendation for this task (2,000 satellite images, 4 crop classes):
  -> Start with Feature Extraction.

Reasons:
  1. Small dataset (2,000 samples): fine-tuning the base model
     immediately risks severe overfitting.
  2. Domain gap exists (satellite vs. natural images), but low-level
     features (edges, colour gradients) are still transferable.
  3. Training only the head first prevents large, noisy gradients
     from a randomly initialised classifier from corrupting the
     pre-trained weights.
  4. Once the head converges, selective fine-tuning of the last few
     base layers can safely improve performance further.
''')
```

```python
print('=' * 60)
print('QUESTION 6(b): Transfer Learning Code Bugs')
print('=' * 60)

print('''
BUGGY CODE:
-----------
base_model = tf.keras.applications.ResNet50(
    weights="imagenet",
    include_top=True,                        # <-- BUG A
    input_shape=(224, 224, 3)
)
base_model.trainable = False
model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.Dense(256, activation="relu"),
    tf.keras.layers.Dense(4, activation="softmax")
])
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-2),  # <-- BUG B
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

BUG A: include_top=True
  PROBLEM:
    include_top=True loads ResNet50 WITH its original ImageNet
    classification head (Flatten + Dense(1000, softmax)).
    Adding Dense(256) and Dense(4) on top of an existing Dense(1000)
    creates an architecture mismatch and makes the model
    architecturally incorrect for the transfer learning setup.

  FIX: include_top=False
    Loads only the convolutional base; we then attach our own
    GlobalAveragePooling2D and Dense layers.

BUG B: Adam(1e-2) -- learning rate far too high
  PROBLEM:
    lr=0.01 is too aggressive for transfer learning.
    Even for the head-only phase it can cause oscillations.
    During fine-tuning, such a high LR would destroy
    the carefully pre-trained base weights.

  FIX: Adam(1e-3) for head training; Adam(1e-5) for fine-tuning.
''')

# Corrected code
print('CORRECTED CODE:')
base_model = tf.keras.applications.ResNet50(
    weights='imagenet',
    include_top=False,        # FIX A
    input_shape=(224, 224, 3)
)
base_model.trainable = False

model_q6 = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(4,   activation='softmax')
])
model_q6.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),   # FIX B
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
trainable_head = sum(tf.size(v).numpy() for v in model_q6.trainable_variables)
print(f'  Model built successfully')
print(f'  Total parameters     : {model_q6.count_params():,}')
print(f'  Trainable parameters : {trainable_head:,}  (new head only, base frozen)')
```

```python
print('=' * 60)
print('QUESTION 6(c): Fine-Tuning -- Unfreeze Last 20 Layers')
print('=' * 60)

# Step 1: Unfreeze the entire base model
base_model.trainable = True

# Step 2: Refreeze all but the last 20 layers
for layer in base_model.layers[:-20]:
    layer.trainable = False

# Step 3: Recompile with a very low learning rate
model_q6.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

trainable_ft  = sum(tf.size(v).numpy() for v in model_q6.trainable_variables)
unfrozen_cnt  = sum(1 for l in base_model.layers if l.trainable)
print(f'  Trainable parameters after fine-tuning setup : {trainable_ft:,}')
print(f'  Unfrozen base model layers                   : {unfrozen_cnt} / {len(base_model.layers)}')

print('''
Code summary:
-------------
  # Step 1
  base_model.trainable = True

  # Step 2
  for layer in base_model.layers[:-20]:
      layer.trainable = False

  # Step 3
  model.compile(
      optimizer=tf.keras.optimizers.Adam(1e-5),
      loss="categorical_crossentropy",
      metrics=["accuracy"]
  )

Why must the classifier head be trained BEFORE unfreezing the base?
  When the new Dense layers are randomly initialised, they produce
  very large, noisy gradients during the first backward passes.
  If the base model were already unfrozen at that point, these
  large gradients would propagate all the way through ResNet50
  and overwrite its carefully learned ImageNet representations --
  a phenomenon known as catastrophic forgetting.

  By first training only the head:
    1. The new layers converge to sensible weights.
    2. Gradient magnitudes become small and stable.
    3. Fine-tuning the base at lr=1e-5 then makes only tiny,
       controlled adjustments, preserving the valuable pre-trained
       representations while adapting them to the satellite domain.
''')
```
